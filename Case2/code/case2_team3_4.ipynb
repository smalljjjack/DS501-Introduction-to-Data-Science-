## Business Objects

Business objects contain basic information about local businesses. The fields are as follows:

```json
{
  'type': 'business',
  'business_id': (a unique identifier for this business),
  'name': (the full business name),
  'neighborhoods': (a list of neighborhood names, might be empty),
  'full_address': (localized address),
  'city': (city),
  'state': (state),
  'latitude': (latitude),
  'longitude': (longitude),
  'stars': (star rating, rounded to half-stars),
  'review_count': (review count),
  'photo_url': (photo url),
  'categories': [(localized category names)]
  'open': (is the business still open for business?),
  'schools': (nearby universities),
  'url': (yelp url)
}
```
## Checkin Objects
```json
{
    'type': 'checkin',
    'business_id': (encrypted business id),
    'checkin_info': {
        '0-0': (number of checkins from 00:00 to 01:00 on all Sundays),
        '1-0': (number of checkins from 01:00 to 02:00 on all Sundays),
        ...
        '14-4': (number of checkins from 14:00 to 15:00 on all Thursdays),
        ...
        '23-6': (number of checkins from 23:00 to 00:00 on all Saturdays)
```

# Data Collection/Processing: 

````python
#
# Enum definition and code to load yelp dataset from multiple files
#
from enum import Enum
import json
import os

# The types of Yelp Data that can be loaded
#
class YelpDataType(Enum):
    BUSINESS = 'business'  #UniqueID attribute: 'business_id'
    USER = 'user'#UniqueID: user_id
    CHECKIN = 'checkin' #Biz Hrs join to Business by attribute: 'business_id' format <day>-<openHr>:<closeHr>
    REVIEW = 'review' #UniqueID: review_id  join to Business by attribute 'business_id', user by attribute 'user_id
    TIP = 'tip' # tip on place, join to Business by attribute 'business_id', user by attribute 'user_id
    def __str__(self):
        return str(self.value)

#
# NOTE: we 'cut' the yelp datasets into files no more than 240 MB
#       this was for easy transfer and box storage
#
# load yelp data that was split in chunks
# ./yelp_<datatype>_<counter>.json
#
def loadYelpData(datatype):
    result = []
    filecount = 1
    while True:
        filename = './yelp_{0}_{1}.json'.format(datatype,filecount)
        if not os.path.isfile(filename):
            return result
        print('  Loading {0}'.format(filename))
        with open(filename, 'r') as f:
            result += json.load(f)
        f.closed
        filecount += 1
    return result

````

````python
YELP_DATA = {}
for datatype in YelpDataType:
    print('Loading Yelp {0} data'.format(datatype))
    YELP_DATA[datatype] = loadYelpData (datatype)
````

**(1) Finding the most popular business categories:** 
* print the top 10 most popular business categories in the dataset and their counts (i.e., how many business objects in each category). Here we say a category is "popular" if there are many business objects in this category (such as 'restaurants').

````python
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary
from collections import Counter
#
# load categories for each business.
# presumes a business only has a category once (not twice)
# returns counter of category array
def loadCategoryCounter(yelpBusinesses):
    categoryArray = []
    for business in yelpBusinesses:
        categories = business['categories']
        if categories:
            for category in business['categories']:
                categoryArray.append(category.lower())
    return Counter(categoryArray)
````

````python
import matplotlib.pyplot as plt
Top10Categories = loadCategoryCounter(YELP_DATA[YelpDataType.BUSINESS]).most_common(10)

fig, ax = plt.subplots(1,1)
ax.axis('tight')
ax.axis('off')
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
theTable = ax.table(cellText=Top10Categories,colLabels=('Category','Business Object Count'), loc='center')
theTable.set_fontsize(25)
theTable.scale(5, 5)
plt.show()
````

** (2) Find the most popular business objects** 
* print the top 10 most popular business objects in the dataset and their counts (i.e., how many checkins in total for each business object).  Here we say a business object is "popular" if the business object attracts a large number of checkins from the users.

````python
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary
from collections import Counter
#
# load businessId as many times as checkin counts exist
# return counter of business ids to rank
def businessIDCheckinCounter(yelpCheckins):
    result = []
    for checkin in yelpCheckins:
        checkinCount = 0
        for checkinTimeCount in checkin['time']:
            checkinCount += int(checkinTimeCount.split(':')[1])
        result += [checkin['business_id']] * checkinCount
    return Counter(result)

#
# convert the array of businesses into
# a business_id -> business map
#
def createBusinessIdMap(yelpBusinesses):
    result = {}
    for business in yelpBusinesses:
        result[business['business_id']] = business
    return result
````

````python
import matplotlib.pyplot as plt
Top10Businesses = businessIDCheckinCounter(YELP_DATA[YelpDataType.CHECKIN]).most_common(10)
businessIdMap = createBusinessIdMap(YELP_DATA[YelpDataType.BUSINESS])

fig, ax = plt.subplots(1,1)
ax.axis('tight')
ax.axis('off')
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
theTable = ax.table(cellText=[(businessIdMap[row[0]]['name'], row[0], row[1]) for row in Top10Businesses],
         colLabels=('Business','BusinessId', 'Checkin Count'), loc='center')
theTable.set_fontsize(25)
theTable.scale(5, 5)
plt.show()
````

````python
# 1) Convert Business, Review and User data into a single collection of User Profiles.
# Aggregating Review and Business data into user profiles
# for collaborative filtering to find 
# suggested new 'friends' and suggested new 'businesses' 
#
import json
import os
import sys
# attribute fields for JSON
USER_ID = 'user_id'
BUSINESS_ID = 'business_id'
POSTAL_CODE = 'postal_code'
CATEGORIES = 'categories'
POSTAL_CODE_MAP = 'postal_code_map'
BUSINESS_MAP = 'business_map'
CATEGORY_MAP = 'category_map'
STARS = 'stars'
REVIEWS = 'reviews'

#
# NOTE: we 'cut' the yelp datasets into files no more than 240 MB
#       this was for easy transfer and box storage
#
# load yelp data that was split in chunks
# ./yelp_<datatype>_<counter>.json
#
def loadYelpData(datatype):
    result = []
    filecount = 1
    while True:
        filename = './yelp_{0}_{1}.json'.format(datatype,filecount)
        if not os.path.isfile(filename):
            return result
        print('  Loading {0}'.format(filename))
        with open(filename, 'r') as f:
            result += json.load(f)
        f.closed
        filecount += 1
    return result

# index passed data by passed keyfield
def indexData(data, keyField):
    result = {}
    for item in data:
        result[item[keyField].lower()] = item
    return result

#
# write  data out into multiple files named like datatype of max size
#
def WriteOutInChunks(data, datatype, maxbytesperfile):
    result = []
    filebytes = 0
    filecount = 1
    filename = './yelp_{0}_{1}.json'.format(datatype, filecount)
    for user in data:
        line = json.dumps(user)
        result.append(user)
        filebytes += len(line)
        if filebytes > maxbytesperfile:
            print('  Writing {0}'.format(filename))
            with open(filename, 'w') as newFile:
                json.dump(result, newFile)
            newFile.closed
            result.clear()
            filebytes = 0
            filecount += 1
            filename = './yelp_{0}_{1}.json'.format(datatype, filecount)
    if 0 < filebytes:
        print('  Writing {0}'.format(filename))
        with open(filename, 'w') as newFile:
            json.dump(result, newFile)
        newFile.closed

#
# combine review data with businessDataMap to
# construct user profile of category (counts), business stars counts
#
def aggregateReviewAndBusinessDataToUserProfile(businessDataMap, reviewData, userDataMap):
    result = {}
    usedReviewCount = 0
    businessesWithoutCategories = {}
    reviewedBusinesses = {}

    for review in reviewData:
        businessId = review[BUSINESS_ID].lower()
        userId = review[USER_ID].lower()
        if not businessId in businessDataMap:
            continue
        if not userId in userDataMap:
            continue
        if not businessId in reviewedBusinesses:
            reviewedBusinesses[businessId] = True
        usedReviewCount += 1
        business = businessDataMap[businessId]
        userId = review[USER_ID].lower()
        postalCode = business[POSTAL_CODE]
        postalCodeCount = 1
        starCount = review[STARS]
        reviewCount = 1
        businessReview = { STARS: starCount, REVIEWS: reviewCount }
        categoryMap = {}
        if not businessId in businessesWithoutCategories:
            try:
                for category in business[CATEGORIES]:
                    categoryMap[category.lower()] = 1
            except:
                businessesWithoutCategories[businessId] = True
        # fill in new userId review data
        if not userId in result:
            result[userId] = { USER_ID: userId,
                               POSTAL_CODE_MAP: { postalCode: postalCodeCount },
                               BUSINESS_MAP: { businessId: businessReview },
                               CATEGORY_MAP: categoryMap
                             }
            continue

        # add to existing userId review data
        userRecord = result[userId]

        # postal codes
        if postalCode in userRecord[POSTAL_CODE_MAP]:
            postalCodeCount += userRecord[POSTAL_CODE_MAP][postalCode]
        userRecord[POSTAL_CODE_MAP][postalCode] = postalCodeCount

        # businesses
        if businessId in userRecord[BUSINESS_MAP]:
            starCount += userRecord[BUSINESS_MAP][businessId][STARS]
            reviewCount += userRecord[BUSINESS_MAP][businessId][REVIEWS]
            businessReview = { STARS: starCount, REVIEWS: reviewCount }
        userRecord[BUSINESS_MAP][businessId] = businessReview

        # categories
        for category in categoryMap.keys():
            categoryCount = 1
            if category in userRecord[CATEGORY_MAP]:
                categoryCount += userRecord[CATEGORY_MAP][category]
            userRecord[CATEGORY_MAP][category] = categoryCount
    
    print('{0}/{1} reviews used {2}/{3} businesses used, {4}/{2} reviewed businesses have no categories {5}/{6} users profiled'.format(
            usedReviewCount, len(reviewData),
        len(reviewedBusinesses), len(businessDataMap),len(businessesWithoutCategories),
        len(result), len(userDataMap)))
    return result
````

````python
# 1) Convert Business, Review and User data into a single collection of User Profiles.
# construct userprofile files
#
businessDataMap = indexData(loadYelpData('business'),BUSINESS_ID)
userDataMap = indexData(loadYelpData('user'),USER_ID)

userProfile = aggregateReviewAndBusinessDataToUserProfile(businessDataMap, loadYelpData('review'), userDataMap)
WriteOutInChunks(userProfile.values(), 'userprofile', 225000000)
````

````python
#2) Using the userprofiles, we construct our per-postal code user to category review count matrices as csv files. 
#
# Cut userprofile data into
# per postal-code csv files
# of userid, category1ReviewCount,......, categoryNReviewCount
import os
import json
# attribute fields for JSON
POSTAL_CODE_MAP = 'postal_code_map'
CATEGORY_MAP = 'category_map'

#
# We need profiles with category data to map new friends 
#
def extractProfilesWithCategories(userprofiles):
    result = []
    for userprofile in userprofiles:
        if 0 < len(userprofile[CATEGORY_MAP]):
            result.append(userprofile)
    print('{0}/{1} userprofiles with categories extracted.'.format(len(result),len(userprofiles)))
    return result
    
#
# collect all categories from userprofiles
# return map(set) of categories
def collectAllCategories(userprofiles):
    categoryMap = {}
    for userprofile in userprofiles:
        for category in userprofile[CATEGORY_MAP].keys():
            if not category in categoryMap:
                categoryMap[category] = True
    return categoryMap.keys()
#
# collect mapping of postalCodes to userIds
#
def collectPostalCodeUserIds(userprofiles):
    postalCodeUserIds = {}
    for userprofile in userprofiles:
        for postalCode in userprofile[POSTAL_CODE_MAP].keys():
            userIdArray = []
            if postalCode in postalCodeUserIds:
                userIdArray = postalCodeUserIds[postalCode]
            userIdArray.append(userprofile[USER_ID])
            postalCodeUserIds[postalCode] = userIdArray
    return postalCodeUserIds
#
# collect mapping of userId to value for ALL categories (each category as an item)
#
def userIdCategoryRating(userprofiles, allCategories):
    userCategoryValues = {}
    for userprofile in userprofiles:
       categoryMap = {}
       for category in allCategories:
           categoryValue = 0
           if category in userprofile[CATEGORY_MAP]:
               categoryValue = userprofile[CATEGORY_MAP][category]
           categoryMap[category] = categoryValue
       userCategoryValues[userprofile[USER_ID]] = categoryMap
    return userCategoryValues

#
# write out 1 csv file per postal code
# table of userid with categories
#
def writePostalCodeUserCategoryCSV(postalCodeUsersMap, allcategories, userCategoryRatingsMap):
    header = '{0},{1}'.format(USER_ID,','.join(allcategories))
    for postalCode in postalCodeUsersMap.keys():
        filename = './yelp_usercategorymatrix_{0}.csv'.format(postalCode)
        print(' Writing {0}'.format(filename))
        with open(filename, 'w') as newFile:
            newFile.write(header + os.linesep)
            for userId in postalCodeUsersMap[postalCode]:
                newLine = [userId]
                for category in userCategoryRatingsMap[userId].keys():
                    newLine.append(str(userCategoryRatingsMap[userId][category]))
                newFile.write(','.join(newLine) + os.linesep)
        newFile.closed

````

````python
#2) Using the userprofiles, we construct our per-postal code user to category review count matrices as csv files. 
# construct csv files for compiling recommendations per zip code
print('loading user profiles with categories (skip if no categories)')
userprofiles = extractProfilesWithCategories(loadYelpData('userprofile'))
print('collect super-set of categories found in user profiles')
allCategories = collectAllCategories(userprofiles)
print('collect map of postal codes to UserIds')
postalCodeUsersMap = collectPostalCodeUserIds(userprofiles)
print('construct map of userId to super-set of categories review-counts')
userCategoryRatingsMap = userIdCategoryRating(userprofiles, allCategories)
print('Create 1 csv file per postal code of userId,category1,...,categoryN review counts')
writePostalCodeUserCategoryCSV(postalCodeUsersMap, allCategories, userCategoryRatingsMap)
````

````python
# 3) choosing zip-code 89109 (Las Vegas, NV) and using a K value of 100.
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary

import os
import csv
import numpy as np
from sklearn.cluster import KMeans
import json

#reference for kmeans usage: http://codereview.stackexchange.com/questions/52029/k-means-clustering-in-python
POSTALCODE_USERCATEGORY_CSV_FILEPREFIX = 'yelp_usercategorymatrix_'
POSTALCODE_FRIENDCLUSTERS_JSON_FILEPREFIX = 'yelp_friendclusters_'
CSV_FILESUFFIX = '.csv'
CAT_REVIEWCOUNT_MATRIX_FIELD = 'categoryReviewMatrix'
USERID_COLUMN_FIELD = 'userIdColumn'
#
# collect list of all postal code csv files with userCategory tables
#
def getPostalCodeUserCategoryCSVs():
    result = {}
    for file in os.listdir('.'):
        if file.startswith(POSTALCODE_USERCATEGORY_CSV_FILEPREFIX) and file.endswith(CSV_FILESUFFIX):
            postalCode = file[len(POSTALCODE_USERCATEGORY_CSV_FILEPREFIX):(-1*len(CSV_FILESUFFIX))]
            result[postalCode] = './{0}'.format(file)
    return result

#
# load csvFilename
# into np array as a matrix
#
def loadCategoryReviewMatrixandUserIdColumn(csvFilename):
    rowData = None
    with open(csvFilename, 'r') as csvf:
        rowData = list(csv.reader(csvf))
    csvf.closed
    userIdColumn = []
    matrixData = []
    for colArray in rowData[1:]:
        userIdColumn.append(colArray[0])
        matrixRow = []
        for col in colArray[1:]:
            matrixRow.append(int(col))
        matrixData.append(matrixRow)
    result = np.array(matrixData)
    print('Constructed {0}x{1} matrix of Category-ReviewCounts, and {0} column of UserIds'.format(len(userIdColumn), len(matrixData[0])))
    return { USERID_COLUMN_FIELD: userIdColumn, CAT_REVIEWCOUNT_MATRIX_FIELD: result }

#
# extract clusters of friends by applying k-means
# fit to categoryMatrix
#
def extractFriendClusters(numClusters, categoryMatrix, userIdColumn):
    result = []
    userCount = len(userIdColumn)
    for i in range(numClusters):
        result.append([])
    print('Applying kmeans to {0}x{1} matrix to {2} clusters'.format(len(userIdColumn), len(categoryMatrix[0]), numClusters))
    clusteredResult = KMeans(n_clusters=numClusters).fit(categoryMatrix)
    for i in range(userCount):
        clusterIndex = int(clusteredResult.labels_[i])
        friendId = userIdColumn[i]
        result[clusterIndex].append(friendId)
    return result
#
# write the friendClusters for the passed postalCode
# to a file for later usage
#
def writeFriendClustersToFile(postalCode, friendClusters):
    filename = './{0}{1}.json'.format(POSTALCODE_FRIENDCLUSTERS_JSON_FILEPREFIX, postalCode)
    print('Writing {0} clusters of friends to {1}'.format(len(friendClusters), filename))
    with open(filename, 'w') as jsonFile:
        json.dump(friendClusters,jsonFile)
    jsonFile.closed
````

````python
# 3) choosing zip-code 89109 (Las Vegas, NV) and using a K value of 100.
# Your code starts here
#   Please add comments or text cells in between to explain the general idea of each block of the code.
#   Please feel free to add more cells below this cell if necessary
K = 100 # 100 clusters in 100K + users: 89109 is Las Vegas, NV
samplePostalCode = '89109'
sampleCityForPostalCode = 'Las Vegas, NV'
print('Load csv file for {0} {1}'.format(sampleCityForPostalCode, samplePostalCode))
# extract postalcode csv
postalCodeUserCategoryCSVMap = getPostalCodeUserCategoryCSVs()
sampleFile = postalCodeUserCategoryCSVMap[samplePostalCode]

print('Construct Category Matrix for K-Means clustering')
sampleMatrixAndUserIdColumn = loadCategoryReviewMatrixandUserIdColumn(sampleFile)
print('Apply K-means clustering to category Matrix extract {0} clusters of userIds as friend clusters'.format(K))
friendClusters = extractFriendClusters(K, sampleMatrixAndUserIdColumn[CAT_REVIEWCOUNT_MATRIX_FIELD],sampleMatrixAndUserIdColumn[USERID_COLUMN_FIELD] )
print('Write friend clusters to file for later processing')
writeFriendClustersToFile(samplePostalCode,friendClusters)
````

````python
#4) New Friend recommendations:
import os
import sys
import json
from collections import Counter
USER_ID = 'user_id'
FRIENDS = 'friends'
USER_DATATYPE = 'user'
FRIENDCLUSTERS_DATATYPE = 'friendclusters'
FRIENDRECOMMENDATION_DATATYPE = 'friendrecommendations'
#
# load yelp data file of specific subpart of specific datatype
#
def loadYelpDataFile(datatype, subPart):
    result = None
    filename = './yelp_{0}_{1}.json'.format(datatype, subPart)
    if not os.path.isfile(filename):
        return result
    print('  Loading {0}'.format(filename))
    with open(filename, 'r') as f:
        result = json.load(f)
    f.closed
    return result

#
# load yelp data files that were split in chunks
# ./yelp_<datatype>_<counter>.json
def loadYelpData(datatype):
    result = []
    filecount = 1
    while True:
        data = loadYelpDataFile(datatype, str(filecount))
        if not data:
            break
        result += data
        filecount += 1
    return result
#
# index passed data by passed keyfield
def indexData(data, keyField):
    result = {}
    for item in data:
        result[item[keyField].lower()] = item
    return result

#
# obtain statisics on friendClusters
# Num Clusters, Total Users across all clusters, size of each cluster
#
def clusterStatistics(friendClusters):
    numClusters = len(friendClusters)
    clusterSizes = []
    totalUsers = 0
    minSize = sys.maxsize
    maxSize = 0
    for cluster in friendClusters:
        clusterSize = len(cluster)
        if clusterSize < minSize:
            minSize = clusterSize
        if maxSize < clusterSize:
            maxSize = clusterSize
        clusterSizes.append(clusterSize)
        totalUsers += clusterSize
    return { 'ClusterCount': numClusters, 'TotalUsers': totalUsers,
             'AveClusterSize': (totalUsers/numClusters), 'MinClusterSize': minSize, 'MaxClusterSize': maxSize,
             'ClusterSizes': clusterSizes }
#
# obtain statistics on recommendations
# total users (with recommendations), total recommendations, aveRecommendations/user
#
def recommendationStatistics(userIdRecommendationMap):
    totalUsers = len(userIdRecommendationMap)
    totalRecommendations = 0
    minRecommendations = sys.maxsize
    maxRecommendations = 0
    for recommendationArray in userIdRecommendationMap.values():
        recommendationCount = len(recommendationArray)
        if recommendationCount < minRecommendations:
            minRecommendations = recommendationCount
        if maxRecommendations < recommendationCount:
            maxRecommendations = recommendationCount
        totalRecommendations += recommendationCount

    return { 'TotalUsers': totalUsers, 'TotalRecommendations': totalRecommendations,
             'AverageRecommendationsPerUser': (totalRecommendations/totalUsers),
             'MinRecommendations': minRecommendations, 'MaxRecommendations': maxRecommendations }


#
# extract map of UserId -> recommendedFriendArray Map based
# on friendClusters of UserIds and a User's existing friends
# Map will only contain users with at least 1 recommended friend
#
def extractUserIdRecommendedFriendsMapFromClusters(userIdUserMap, friendClusters):
    result = {}
    for friendCluster in friendClusters:
        friendUserIdSet = set(friendCluster)
        for userId in friendCluster:
            user = userIdUserMap[userId.lower()]
            existingFriends = set()
            existingFriends.add(userId) # you should not be your own recommended friend
            if FRIENDS in user:
                existingFriends.update(user[FRIENDS])
            recommendedFriends = friendUserIdSet - existingFriends # set difference
            if 0 < len(recommendedFriends):
                result[userId.lower()] = list(recommendedFriends)
    return result

#
# extract the top X userIds from the userID RecommendationsMap
#
def extractTopXUserIdList(userIdList, userIdUserMap, topCount):
    result = []
    for userId in userIdList:
        user = userIdUserMap[userId.lower()]
        reviewCount = int(user['review_count'])
        result += [userId] * reviewCount
    topX = Counter(result).most_common(topCount)
    return [usercount[0] for usercount in topX]
#
# extract top X users with their Top X recommended friends
#
def extractTopXUsersWithTheirTopXFriends(userIdRecommendedFriends, userIdUserMap, topCount):
    result = {}
    topXUsers = extractTopXUserIdList(userIdRecommendedFriends, userIdUserMap, topCount)
    for userId in topXUsers:
        result[userId] = extractTopXUserIdList(userIdRecommendedFriends[userId], userIdUserMap, topCount)
    return result


##
# write Recommendations to file
#
def writeRecommendations(userIdRecommendationMap, datatype, postalCode, recommendationStats):
    filename = './yelp_{0}_{1}.json'.format(datatype, postalCode)
    print('Writing {0}  {1}'.format(json.dumps(recommendationStats), filename))
    with open(filename, 'w') as jsonFile:
        json.dump(userIdRecommendationMap,jsonFile)
    jsonFile.closed
````

````python
#4) New Friend recommendations:
postalCode = '89109'
postalCodeTownName = 'Las Vegas, NV'
print('Loading Yelp Users, and friendClusters for {0} {1}'.format(postalCodeTownName, postalCode))
userIdUserMap       = indexData(loadYelpData(USER_DATATYPE), USER_ID)
clusteredUsers = loadYelpDataFile(FRIENDCLUSTERS_DATATYPE, postalCode)
clusterStats = clusterStatistics(clusteredUsers)
print()
print('Extracting Recommended Friends for users in {0} {1} for cluster {2}'.format(postalCodeTownName, postalCode, json.dumps(clusterStats)))

userIdRecommendedFriendsMap = extractUserIdRecommendedFriendsMapFromClusters(userIdUserMap, clusteredUsers)
recommendedFriendsStats = recommendationStatistics(userIdRecommendedFriendsMap)
print()
print('Recommended Friends results: {0}  for {1} {2}'.format(json.dumps(recommendedFriendsStats), postalCodeTownName, postalCode))
print()
print('Cutting back listing to Top3 users and their Top3 recommended friends for {0} {1}'.format(postalCodeTownName, postalCode))
top3UserIDRecommendedFriendsMap = extractTopXUsersWithTheirTopXFriends(userIdRecommendedFriendsMap, userIdUserMap, 3)
top3recommendedFriendsStats = recommendationStatistics(top3UserIDRecommendedFriendsMap)
print()
print('Top3 Recommended Friends results: {0}  for {1} {2}'.format(json.dumps(top3recommendedFriendsStats), postalCodeTownName, postalCode))
print()
writeRecommendations(top3UserIDRecommendedFriendsMap, FRIENDRECOMMENDATION_DATATYPE, postalCode, top3recommendedFriendsStats)
````

````python
#5) New Business recommendations:
import json
import os
import sys
from collections import Counter
USER_ID = 'user_id'
BUSINESS_MAP ='business_map'
BUSINESS_RATING_SUMTOTAL='stars'
BUSINESS_REVIEW_COUNT = 'reviews'
USERPROFILE_DATATYPE = 'userprofile'
FRIENDRECOMMENDATION_DATATYPE = 'friendrecommendations'
BUSINESSRECOMMENDATION_DATATYPE = 'businessrecommendations'
#
# load yelp data file of specific subpart of specific datatype
#
def loadYelpDataFile(datatype, subPart):
    result = None
    filename = './yelp_{0}_{1}.json'.format(datatype, subPart)
    if not os.path.isfile(filename):
        return result
    print('  Loading {0}'.format(filename))
    with open(filename, 'r') as f:
        result = json.load(f)
    f.closed
    return result

#
# load yelp data files that were split in chunks
# ./yelp_<datatype>_<counter>.json
def loadYelpData(datatype):
    result = []
    filecount = 1
    while True:
        data = loadYelpDataFile(datatype, str(filecount))
        if not data:
            break
        result += data
        filecount += 1
    return result
#
# index passed data by passed keyfield
def indexData(data, keyField):
    result = {}
    for item in data:
        result[item[keyField].lower()] = item
    return result

# obtain statistics on recommendations
# total users (with recommendations), total recommendations, aveRecommendations/user
#
def recommendationStatistics(userIdRecommendationMap):
    totalUsers = len(userIdRecommendationMap)
    totalRecommendations = 0
    minRecommendations = sys.maxsize
    maxRecommendations = 0
    for recommendationArray in userIdRecommendationMap.values():
        recommendationCount = len(recommendationArray)
        if recommendationCount < minRecommendations:
            minRecommendations = recommendationCount
        if maxRecommendations < recommendationCount:
            maxRecommendations = recommendationCount
        totalRecommendations += recommendationCount

    return { 'TotalUsers': totalUsers, 'TotalRecommendations': totalRecommendations,
             'AverageRecommendationsPerUser': (totalRecommendations/totalUsers),
             'MinRecommendations': minRecommendations, 'MaxRecommendations': maxRecommendations }


#
# extract Reviewed Business Ids where average review rating was >= the provided minimumStarRating
#
def extractReviewedBusinessIdMap(userProfileBusinessMap, minimumStarRating):
    result = {}
    for businessId in userProfileBusinessMap.keys():
        ratingTotal = int(userProfileBusinessMap[businessId][BUSINESS_RATING_SUMTOTAL])
        reviewCount = int(userProfileBusinessMap[businessId][BUSINESS_REVIEW_COUNT])
        aveRating = ratingTotal/reviewCount
        if minimumStarRating <= aveRating:
            result[businessId] = userProfileBusinessMap[businessId]
    return result
#
# extract lhsMap - rhsMap (similar to Set difference operation
#
def extractDifference(lhsMap, rhsMap):
    result = {}
    lhsKeys = set(lhsMap.keys())
    rhsKeys = set(rhsMap.keys())
    lhsSubsetKeys = lhsKeys - rhsKeys
    for key in lhsSubsetKeys:
        result[key] = lhsMap[key]
    return result
#
# extract map of UserId -> recommendedBusinessesArray } based
# on userId -> recommendedFriends and minimum review rating for recommended businesses
# Map will only contain users with at least 1 recommended business
#
def extractUserIdRecommendedBusinessesFromRecommendedFriends(userIdUserProfileMap, userIdRecommendedFriendsMap, minimumStarRating):
    result = {}
    for userId in userIdRecommendedFriendsMap.keys():
        userProfile = userIdUserProfileMap[userId]
        recommendedFriends = userIdRecommendedFriendsMap[userId]
        allExistingBusinesses  = extractReviewedBusinessIdMap(userProfile[BUSINESS_MAP], 0)
        allRecommendedBusinesses = {}
        for newFriendUserId in recommendedFriends:
            friendProfile = userIdUserProfileMap[newFriendUserId.lower()]
            newBusinesses = extractReviewedBusinessIdMap(friendProfile[BUSINESS_MAP], minimumStarRating)
            newRecommendedBusinesses = extractDifference(extractDifference(newBusinesses, allRecommendedBusinesses), allExistingBusinesses)
            allRecommendedBusinesses = newRecommendedBusinesses
        if 0 < len(allRecommendedBusinesses):
            result[userId] = allRecommendedBusinesses
    return result
#
# extractUserIdTopXRecommendedBusinesses
# TopCount recommended businesses for each user
def extractUserIdTopXRecommendedBusinesses(userIdBusinessRecommendationsMap, topCount):
    result = {}
    for userId in userIdBusinessRecommendationsMap.keys():
        businessRecommendationsMap = userIdBusinessRecommendationsMap[userId]
        counterArray = []
        for businessId in businessRecommendationsMap.keys():
            business = businessRecommendationsMap[businessId]
            counterArray += [businessId] * int(int(business[BUSINESS_RATING_SUMTOTAL])/int(business[BUSINESS_REVIEW_COUNT]))
        topX = Counter(counterArray).most_common(topCount)
        newRecommendations = {}
        for businessIdCount in topX:
            newRecommendations[businessIdCount[0]] = businessRecommendationsMap[businessIdCount[0]];
        result[userId] = newRecommendations
    return result
#
# normalize userId -> businessIdBusinessMap to userId -> businessIdlist
#
def normalizeUserIdRecommendedBusinesses(userIdBusinessRecommendationsMap):
    result = {}
    for userId in userIdBusinessRecommendationsMap.keys():
        result[userId] = list(userIdBusinessRecommendationsMap[userId].keys())
    return result
#
# write Recommendations to file
#
def writeRecommendations(userIdRecommendationMap, datatype, postalCode, recommendationStats):
    filename = './yelp_{0}_{1}.json'.format(datatype, postalCode)
    print('Writing {0}  {1}'.format(json.dumps(recommendationStats), filename))
    with open(filename, 'w') as jsonFile:
        json.dump(userIdRecommendationMap,jsonFile)
    jsonFile.closed
````

````python
#5) New Business recommendations:

minimumStarRating = 3
postalCode = '89109'
postalCodeTownName = 'Las Vegas, NV'
print('Loading Yelp UserProfiles, and RecommendedFriends File for {0} {1}'.format(postalCodeTownName, postalCode))
userIdUserProfileMap = indexData(loadYelpData(USERPROFILE_DATATYPE), USER_ID)
recommendedFriends = loadYelpDataFile(FRIENDRECOMMENDATION_DATATYPE, postalCode)

recommendedFriendsStats = recommendationStatistics(recommendedFriends)
print()
print('Recommended Friends results: {0}  for {1} {2}'.format(json.dumps(recommendedFriendsStats), postalCodeTownName, postalCode))
print()
print('Extracting Recommended Businesses with minimum Ave Star Rating of {0} for Recommended Friends {1}.'.format(minimumStarRating, json.dumps(recommendedFriendsStats)))
recommendedBusinesses = extractUserIdRecommendedBusinessesFromRecommendedFriends(userIdUserProfileMap, recommendedFriends, minimumStarRating)
recommendedBusinessStats = recommendationStatistics(recommendedBusinesses)
recommendedBusinessStats['MinimumAveStarRating'] = minimumStarRating
print()
print('Recommended Business results: {0}  for {1} {2}'.format(json.dumps(recommendedBusinessStats), postalCodeTownName, postalCode))
print()
print('Cutting back listing to Top3 recommended businesses for {0} {1}'.format(postalCodeTownName, postalCode))
top3UserIDRecommendedBusinessMap = normalizeUserIdRecommendedBusinesses(extractUserIdTopXRecommendedBusinesses(recommendedBusinesses, 3))
top3recommendedBusinessesStats = recommendationStatistics(top3UserIDRecommendedBusinessMap)
print()
print('Top3 Recommended businesses results: {0}  for {1} {2}'.format(json.dumps(top3recommendedBusinessesStats), postalCodeTownName, postalCode))
print()
writeRecommendations(top3UserIDRecommendedBusinessMap, BUSINESSRECOMMENDATION_DATATYPE, postalCode, top3recommendedBusinessesStats)
````

````python
#6) Display our results for the top 3 users. If the number of friend recommendations is too large, 
#   we will only render the top 3 recommended friends for the top 3 users.
import os
import json
from collections import Counter
USER_ID = 'user_id'
BUSINESS_ID = 'business_id'
USER_DATATYPE = 'user'
BUSINESS_DATATYPE = 'business'
USERPROFILE_DATATYPE = 'userprofile'
FRIENDRECOMMENDATION_DATATYPE = 'friendrecommendations'
BUSINESSRECOMMENDATION_DATATYPE = 'businessrecommendations'
#
# load yelp data file of specific subpart of specific datatype
#
def loadYelpDataFile(datatype, subPart):
    result = None
    filename = './yelp_{0}_{1}.json'.format(datatype, subPart)
    if not os.path.isfile(filename):
        return result
    print('  Loading {0}'.format(filename))
    with open(filename, 'r') as f:
        result = json.load(f)
    f.closed
    return result
#
# load yelp data files that were split in chunks
# ./yelp_<datatype>_<counter>.json
def loadYelpData(datatype):
    result = []
    filecount = 1
    while True:
        data = loadYelpDataFile(datatype, str(filecount))
        if not data:
            break
        result += data
        filecount += 1
    return result
#
# index passed data by passed keyfield
def indexData(data, keyField):
    result = {}
    for item in data:
        result[item[keyField].lower()] = item
    return result
#
# Obtain Renderable Text describing user
#
def getUserText(userId, userIdUserMap, userIdProfileMap, newLine):
    user = userIdUserMap[userId.lower()]
    profile = userIdProfileMap[userId.lower()]
    userName = user['name']
    reviewCount = user['review_count']
    userFriendCount = len(user['friends'])
    userInterests = [] 
    # compile list of interests in order of strongest -> weakest
    for interest in profile['category_map'].keys():
        printableInterest = interest.replace(' & ','-').replace(' (','(').replace(' ','_')
        userInterests += [printableInterest]*int(profile['category_map'][interest])
    Top5Interests = Counter(userInterests).most_common(5)
    userInterests = ['{0}({1})'.format(interest[0],interest[1]) for interest in Top5Interests]
    return '{0} (Id: {1}){6}{2} reviews{6}{3} friends{6}Top 5/{4} Interests(ReviewCount):{6}{5}'.format(userName, userId,
                                                        reviewCount,
                                                        userFriendCount,
                                                        len(profile['category_map']),
                                                        ', '.join(userInterests),
                                                        newLine)
#
# Obtain Renderable Text describing business
#
def getBusinessText(businessId, businessIdBusinessMap, newLine):
    business = businessIdBusinessMap[businessId.lower()]
    bizName = business['name']
    bizAddress = business['address']
    bizCity = business['city']
    bizState = business['state']
    bizZip   = business['postal_code']
    bizRating= business['stars']
    bizReviewCount = business['review_count']
    renderableCategories = []
    for cat in business['categories']:
        renderableCategories.append(cat.lower().replace(' & ','-').replace(' (','(').replace(' ','_'))
    categories = ', '.join(renderableCategories)

    return '{0} (Id: {1}){9}{2}, {3}, {4} {5}{9}{6} stars {7} reviews{9}Categories:{9}{8}'.format(bizName, businessId,
                                                                                         bizAddress, bizCity, bizState, bizZip,
                                                                                         bizRating, bizReviewCount,
                                                                                         categories,
                                                                                         newLine)

postalCode = '89109'
postalCodeTownName = 'Las Vegas, NV'
print('Loading yelp users, yelp businesses, compiled userprofiles,\nuser friend recommendations, and user business recommendations for {0} {1}'.format(postalCodeTownName, postalCode))
userIdUserMap = indexData(loadYelpData(USER_DATATYPE), USER_ID)
userIdProfileMap = indexData(loadYelpData(USERPROFILE_DATATYPE), USER_ID)
businessIdBusinessMap = indexData(loadYelpData(BUSINESS_DATATYPE), BUSINESS_ID)
userIdFriendRecommendations = loadYelpDataFile(FRIENDRECOMMENDATION_DATATYPE, postalCode)
userIdBusinessRecommendations = loadYelpDataFile(BUSINESSRECOMMENDATION_DATATYPE, postalCode)
````
